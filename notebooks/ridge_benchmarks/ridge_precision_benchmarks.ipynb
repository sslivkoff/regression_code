# precision benchmarks on ridge solvers

```python
from __future__ import print_function

import numpy as np

from regression_code.storm.tests import test_datasets
from regression_code.storm.tests import test_utils
from regression_code.storm.ridge.tests import test_ridge
```

# solve_ridge() float32 precision

```python
datasets = test_datasets.load_all(dtype=np.float32)
test_utils.rprint(datasets)
```

```python
common = {
    'weights': False,
    'predictions': False,
    'performance': True,
    'ridges': np.logspace(0, 4, 2),
    'verbose': False,
    'Ytest_zscored': True,
}

results = test_utils.benchmark_functions(
    functions=test_ridge.ridge_functions,
    common=common,
    datasets=datasets,
    outer_loops=1,
    inner_loops=1,
    save_outputs=True,
)
```

```python
test_utils.print_output_precision(
    results,
    baseline_function='aone.models.solve_l2_primal',
    keys=None,
)
```
